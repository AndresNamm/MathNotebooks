```python
# prompt: please generate me code for sample linear transformation moving 3 points around . Make 3 plots 1 befor, 2 after 1 linear transformation, 3 after 3 linear transformation. Make the initial points in some structure. . cOLOR THESE POINTS ALSO

import matplotlib.pyplot as plt
import numpy as np


def point_mover(point, transformation_matrix):
    transformed_point = np.dot(transformation_matrix, point.T)
    return transformed_point

def present_example(point,transformation_matrix):
    transformed_point = point_mover(point, transformation_matrix)
    plt.scatter(point[0], point[1], color='red', label='Initial Point')
    plt.scatter(transformed_point[0], transformed_point[1], color='blue', label='Transformed Point')
    plt.xlim(0, 10)
    plt.ylim(0, 10)
    plt.title("Initial and Transformed Points")
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.show()

# Initial point

point = np.array([1,1])

# Transformation matrix
transformation_matrix = np.array([
  [2, 0],
  [0, 2]]
)


present_example(point,transformation_matrix)


```

```python
point = np.array([1,1])

# Transformation matrix
transformation_matrix = np.array([
  [1, 0],
  [1, 1]]
)


present_example(point,transformation_matrix)
```

```python
# Function to apply a linear transformation
def apply_transformation(points, matrix):
    transformed_points = np.dot(points, matrix.T)  # Correctly apply the matrix
    return transformed_points

# Apply the transformation once and thrice

transformed_points_once = apply_transformation(points, transformation_matrix)

transformed_points_thrice = points.copy()
for _ in range(3):
  transformed_points_thrice = apply_transformation(transformed_points_thrice,transformation_matrix)

# Plotting

plt.figure(figsize=(15, 5))


# Plot initial points
plt.subplot(1, 3, 1)
plt.scatter(points[:, 0], points[:, 1], color='red', label='Initial Points')
plt.title("Initial Points")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

#Plot after first transformation
plt.subplot(1, 3, 2)
plt.scatter(transformed_points_once[:, 0], transformed_points_once[:, 1], color='blue', label='Transformed Points (Once)')
plt.title("Points after one Transformation")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()


# Plot after three transformations
plt.subplot(1, 3, 3)
plt.scatter(transformed_points_thrice[:, 0], transformed_points_thrice[:, 1], color='green', label='Transformed Points (Thrice)')
plt.title("Points after three Transformations")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.tight_layout()
plt.show()
```
